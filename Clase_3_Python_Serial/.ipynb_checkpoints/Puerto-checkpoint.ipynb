```python
ser = serial.serial_for_url('loop://', timeout=1)
ser.isOpen()
ser.timeout=None
ser.flushInput()
ser.flushOutput()
#format(ord(i),"08b")

dato= input('1Calculadora,2Graficar,exit')
while(dato !='1' and dato !='2' and dato !='exit'):
            print('Comando invalido')
            dato=input('1- Calculadora,2-Graficar,exit')

                       
out=''
ser.write(dato.encode())
while ser.inWaiting() > 0 :
    out+=ser.read(1).decode()
print('out:',out)

if(out=="exit"):
    ser.close()
    print('cerrado')                  
                       
if(out=="1"):
     Calculadora.calculadora()
        
if(out=="2"):
     Grafico.graficar()
```
